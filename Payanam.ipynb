````python
from bs4 import BeautifulSoup
import re

def clean_html(text):
    """Removes HTML tags like <br> and <strong> and formats text nicely."""
    # Replace <br> tags with newlines for better formatting
    text = text.replace("<br><br>", "\n\n").replace("<br>", "\n")

    # Use BeautifulSoup to remove all HTML tags safely
    soup = BeautifulSoup(text, "html.parser")
    clean_text = soup.get_text()

    # Remove excessive newlines or spaces
    clean_text = re.sub(r'\n\s*\n', '\n\n', clean_text.strip())

    return clean_text
````

````python
from typing import List, Dict, Any
import random

class FlightTools:
    """Collection of tools available to the agent with expanded dataset"""

    # 50 Major cities/locations
    CITIES = [
        "New York", "London", "Paris", "Tokyo", "Dubai", "Singapore", "Los Angeles",
        "Hong Kong", "Barcelona", "Amsterdam", "Rome", "Madrid", "Sydney", "Miami",
        "Bangkok", "Istanbul", "San Francisco", "Chicago", "Toronto", "Munich",
        "Seoul", "Shanghai", "Beijing", "Delhi", "Mumbai", "Mexico City", "Sao Paulo",
        "Buenos Aires", "Lisbon", "Athens", "Prague", "Vienna", "Copenhagen", "Stockholm",
        "Oslo", "Helsinki", "Dublin", "Edinburgh", "Manchester", "Berlin", "Hamburg",
        "Frankfurt", "Zurich", "Geneva", "Brussels", "Milan", "Venice", "Florence",
        "Las Vegas", "Orlando"
    ]

    AIRLINES = [
        "American Airlines", "Delta", "United Airlines", "British Airways", "Air France",
        "Lufthansa", "Emirates", "Singapore Airlines", "Cathay Pacific", "Qantas",
        "KLM", "Iberia", "Air Canada", "Turkish Airlines", "ANA", "JAL", "Virgin Atlantic",
        "Swiss Air", "Austrian Airlines", "SAS", "Finnair", "Aer Lingus", "Alitalia",
        "TAP Portugal", "EasyJet", "Ryanair", "JetBlue", "Southwest", "Spirit Airlines",
        "Norwegian Air"
    ]

    @staticmethod
    def search_flights(origin: str, destination: str, date: str, passengers: int) -> List[Dict]:
        """Tool 1: Search for available flights - Now with 100+ flight options"""

        # Generate realistic flight data
        flights = []
        num_flights = random.randint(8, 15)  # 8-15 flights per route

        # Calculate base price based on distance (simulated)
        origin_clean = origin.lower().replace(' ', '')
        dest_clean = destination.lower().replace(' ', '')

        # Base price formula (simulated based on route popularity)
        base_price = 200 + (hash(origin_clean + dest_clean) % 500)

        for i in range(num_flights):
            airline = random.choice(FlightTools.AIRLINES)

            # Price variation
            price_variation = random.randint(-100, 300)
            price = base_price + price_variation

            # Departure times
            hour = random.randint(0, 23)
            minute = random.choice([0, 15, 30, 45])
            departure = f"{hour:02d}:{minute:02d}"

            # Duration (4-16 hours)
            duration_hours = random.randint(4, 16)
            duration_minutes = random.choice([0, 15, 30, 45])
            duration = f"{duration_hours}h {duration_minutes}m"

            # Calculate arrival
            arrival_hour = (hour + duration_hours) % 24
            arrival_minute = (minute + duration_minutes) % 60
            day_offset = "+1" if (hour + duration_hours) >= 24 else ""
            arrival = f"{arrival_hour:02d}:{arrival_minute:02d} {day_offset}".strip()

            # Stops
            stops_num = random.choices([0, 1, 2], weights=[60, 30, 10])[0]
            if stops_num == 0:
                stops = "Nonstop"
            elif stops_num == 1:
                stops = "1 stop"
            else:
                stops = "2 stops"

            # Provider/Booking site
            provider = random.choice([
                "Expedia", "Booking.com", "Kayak", "Priceline", "Skyscanner",
                "Direct from Airline", "Google Flights", "Hopper"
            ])

            flights.append({
                "airline": airline,
                "price": price,
                "departure_time": departure,
                "arrival_time": arrival,
                "duration": duration,
                "stops": stops_num,
                "stops_text": stops,
                "provider": provider,
                "flight_number": f"{airline[:2].upper()}{random.randint(100, 9999)}",
                "aircraft": random.choice(["Boeing 737", "Boeing 777", "Airbus A320", "Airbus A380", "Boeing 787"]),
                "class": "Economy"
            })

        # Sort by price
        flights.sort(key=lambda x: x['price'])

        return flights

    @staticmethod
    def analyze_prices(flights: List[Dict]) -> Dict[str, Any]:
        """Tool 2: Analyze flight prices and provide recommendations"""
        if not flights:
            return {}

        prices = [f["price"] for f in flights]
        avg_price = sum(prices) / len(prices)
        min_price = min(prices)
        max_price = max(prices)

        return {
            "average": round(avg_price, 2),
            "min": min_price,
            "max": max_price,
            "recommendation": "Excellent deal!" if min_price < avg_price * 0.85 else "Fair price",
            "trend": random.choice([
                "Prices are 12% lower than last month",
                "Prices are stable compared to last month",
                "Prices are 8% higher than last month"
            ]),
            "best_time": random.choice([
                "Book now - prices expected to rise 8% next week",
                "Good time to book - prices are stable",
                "Wait a few days - prices may drop 5%"
            ])
        }

    @staticmethod
    def get_route_info(origin: str, destination: str) -> Dict[str, str]:
        """Tool 3: Get detailed route information"""
        # Generate realistic route info
        distance = random.randint(500, 8000)
        avg_duration = distance // 500  # Rough estimate: 500 miles per hour

        return {
            "distance": f"{distance:,} miles",
            "timezone": random.choice([
                "EST to GMT (+5 hours)",
                "PST to CET (+9 hours)",
                "EST to JST (+14 hours)",
                "No timezone change"
            ]),
            "peak_season": random.choice([
                "June-August", "December-January", "May-September", "November-February"
            ]),
            "off_peak": random.choice([
                "January-March", "September-November", "February-April", "October-December"
            ]),
            "avg_duration": f"{avg_duration}-{avg_duration+2} hours"
        }

    @staticmethod
    def suggest_alternatives(origin: str, destination: str) -> Dict[str, List]:
        """Tool 4: Suggest alternative airports and dates"""
        # Get some random alternative cities
        alt_origins = random.sample([c for c in FlightTools.CITIES if c.lower() != origin.lower()], 2)
        alt_dests = random.sample([c for c in FlightTools.CITIES if c.lower() != destination.lower()], 2)

        return {
            "nearby_origins": [
                f"{alt_origins[0]} - ${random.randint(20, 80)} cheaper",
                f"{alt_origins[1]} - ${random.randint(20, 80)} cheaper"
            ],
            "nearby_destinations": [
                f"{alt_dests[0]} - ${random.randint(30, 120)} cheaper",
                f"{alt_dests[1]} - ${random.randint(30, 120)} cheaper"
            ],
            "alternate_dates": [
                {"date": "1 day earlier", "savings": f"${random.randint(10, 50)}"},
                {"date": "1 day later", "savings": f"${random.randint(10, 50)}"},
                {"date": "3 days earlier", "savings": f"${random.randint(20, 80)}"}
            ]
        }

    @staticmethod
    def search_hotels(destination: str, check_in: str, check_out: str, guests: int = 2) -> List[Dict]:
        """Tool 5: Search for available hotels - Now with 100+ hotel options"""

        hotels = []
        num_hotels = random.randint(12, 20)  # 12-20 hotels per destination

        hotel_types = [
            "Hotel", "Resort", "Inn", "Suites", "Lodge", "Palace", "Grand Hotel",
            "Boutique Hotel", "Hostel", "Apartment Hotel", "Bed & Breakfast"
        ]

        hotel_names = [
            "Royal", "Grand", "Imperial", "Ritz", "Marriott", "Hilton", "Hyatt",
            "InterContinental", "Sheraton", "Radisson", "Four Seasons", "Westin",
            "Renaissance", "Crowne Plaza", "Holiday Inn", "Best Western", "Comfort",
            "Quality", "Budget", "Economy", "Luxury", "Premier", "Elite", "Central",
            "Downtown", "Riverside", "Oceanview", "Mountain", "Garden", "Plaza"
        ]

        for i in range(num_hotels):
            name_part1 = random.choice(hotel_names)
            name_part2 = random.choice(hotel_types)
            hotel_name = f"{destination} {name_part1} {name_part2}"

            # Price tiers
            tier = random.choices(
                ["budget", "mid", "luxury", "ultra"],
                weights=[30, 40, 20, 10]
            )[0]

            if tier == "budget":
                price = random.randint(50, 100)
                rating = round(random.uniform(3.0, 4.0), 1)
                amenities = random.sample(["WiFi", "Breakfast", "Parking"], k=random.randint(1, 2))
            elif tier == "mid":
                price = random.randint(100, 200)
                rating = round(random.uniform(4.0, 4.5), 1)
                amenities = random.sample(["WiFi", "Breakfast", "Pool", "Gym", "Parking", "Restaurant"], k=random.randint(3, 4))
            elif tier == "luxury":
                price = random.randint(200, 400)
                rating = round(random.uniform(4.5, 4.8), 1)
                amenities = random.sample(["WiFi", "Breakfast", "Pool", "Spa", "Gym", "Restaurant", "Bar", "Concierge", "Room Service"], k=random.randint(5, 7))
            else:  # ultra
                price = random.randint(400, 800)
                rating = round(random.uniform(4.8, 5.0), 1)
                amenities = ["WiFi", "Breakfast", "Pool", "Spa", "Gym", "Fine Dining", "Butler Service", "Private Beach", "Helipad"]

            distance = round(random.uniform(0.1, 8.0), 1)

            hotels.append({
                "name": hotel_name,
                "price_per_night": price,
                "rating": rating,
                "amenities": amenities,
                "distance_from_center": f"{distance} km",
                "room_type": random.choice([
                    "Standard Room", "Deluxe Room", "Premium Room",
                    "Suite", "Junior Suite", "Executive Suite", "Royal Suite"
                ]),
                "provider": random.choice([
                    "Booking.com", "Hotels.com", "Expedia", "Agoda",
                    "Priceline", "Direct Booking", "Trivago"
                ])
            })

        # Sort by price
        hotels.sort(key=lambda x: x['price_per_night'])

        return hotels

    @staticmethod
    def analyze_hotel_prices(hotels: List[Dict]) -> Dict[str, Any]:
        """Tool 6: Analyze hotel prices and provide recommendations"""
        if not hotels:
            return {}

        prices = [h["price_per_night"] for h in hotels]
        ratings = [h["rating"] for h in hotels]

        value_scores = []
        for hotel in hotels:
            value_score = hotel["rating"] / (hotel["price_per_night"] / 100)
            value_scores.append({
                "hotel": hotel["name"],
                "value_score": round(value_score, 2),
                "price": hotel["price_per_night"],
                "rating": hotel["rating"]
            })

        best_value = max(value_scores, key=lambda x: x["value_score"])

        return {
            "min_price": min(prices),
            "max_price": max(prices),
            "average_price": round(sum(prices) / len(prices), 2),
            "highest_rated": max(hotels, key=lambda x: x["rating"])["name"],
            "avg_rating": round(sum(ratings) / len(ratings), 2),
            "best_value": best_value["hotel"],
            "best_value_score": best_value["value_score"],
            "recommendation": f"Best value: {best_value['hotel']} - Great rating ({best_value['rating']}★) at ${best_value['price']}/night"
        }

    @staticmethod
    def get_hotel_package(origin: str, destination: str, date: str, check_in: str,
                          check_out: str, passengers: int = 1) -> Dict[str, Any]:
        """Tool 7: Get complete travel package (flights + hotels)"""
        flights = FlightTools.search_flights(origin, destination, date, passengers)
        hotels = FlightTools.search_hotels(destination, check_in, check_out, passengers)

        flight_analysis = FlightTools.analyze_prices(flights) if flights else {}
        hotel_analysis = FlightTools.analyze_hotel_prices(hotels) if hotels else {}

        if flights and hotels:
            cheapest_flight = min([f["price"] for f in flights])
            cheapest_hotel = min([h["price_per_night"] for h in hotels])
            nights = 3
            package_price = cheapest_flight + (cheapest_hotel * nights)

            return {
                "flights_found": len(flights),
                "hotels_found": len(hotels),
                "cheapest_flight_price": cheapest_flight,
                "cheapest_hotel_per_night": cheapest_hotel,
                "estimated_nights": nights,
                "total_package_price": package_price,
                "flight_analysis": flight_analysis,
                "hotel_analysis": hotel_analysis,
                "savings_tip": f"Book together and save! Package deal: ${package_price} total"
            }

        return {"error": "Unable to create package - flights or hotels not found"}

# Test the tools
tools = FlightTools()
print("✅ Flight Tools created successfully!")
print(f"   - Available cities: {len(FlightTools.CITIES)} locations")
print(f"   - Available airlines: {len(FlightTools.AIRLINES)} carriers")
print(f"   - Testing flights: {len(tools.search_flights('New York', 'London', '2025-12-14', 1))} found")
print(f"   - Testing hotels: {len(tools.search_hotels('London', '2025-12-14', '2025-12-17', 2))} found")
````

````python
from typing import TypedDict, Annotated, List, Dict, Any
import operator
import json

class AgentState(TypedDict):
    """State that flows through the agent graph"""
    query: str
    intent: str
    extracted_params: Dict[str, Any]
    tools_used: Annotated[List[str], operator.add]
    _tools_plan: List[str]
    flights: List[Dict[str, Any]]
    price_analysis: Dict[str, Any]
    route_info: Dict[str, Any]
    alternatives: Dict[str, Any]
    hotels: List[Dict[str, Any]]
    hotel_analysis: Dict[str, Any]
    package_info: Dict[str, Any]
    final_response: str
    messages: Annotated[List, operator.add]

print(" Agent State defined!")


class FlightAgent:
    def __init__(self, llm):
        self.llm = llm
        self.tools = FlightTools()

    def classify_intent(self, state: AgentState) -> AgentState:
        """Node 1: Use OPENAI to classify user intent"""
        system_prompt = """You are an intent classification expert for a travel booking system.
        Classify the user query into one of these intents:
        - simple_search: Basic flight search
        - price_focused: User wants price analysis and deals
        - explore_alternatives: User wants to see alternative options
        - deep_analysis: User wants comprehensive analysis with all information
        - hotel_search: User wants to search for hotels only
        - package_deal: User wants both flights and hotels together

        Respond with ONLY the intent name in lowercase, nothing else."""

        prompt = f"{system_prompt}\n\nUser query: {state['query']}\n\nIntent:"

        try:
            response = self.llm.invoke(prompt)
            intent = response.content.strip().lower().replace('.', '').replace(',', '')

            valid_intents = ["simple_search", "price_focused", "explore_alternatives",
                           "deep_analysis", "hotel_search", "package_deal"]

            if intent not in valid_intents:
                query_lower = state["query"].lower()

                if ("hotel" in query_lower or "accommodation" in query_lower or "stay" in query_lower) and "flight" not in query_lower:
                    intent = "hotel_search"
                elif ("hotel" in query_lower and "flight" in query_lower) or "package" in query_lower or "everything" in query_lower:
                    intent = "package_deal"
                elif "cheap" in query_lower or "best price" in query_lower:
                    intent = "price_focused"
                elif "alternative" in query_lower or "nearby" in query_lower:
                    intent = "explore_alternatives"
                elif any(word in query_lower for word in ["analyze", "compare", "complete", "full", "deep", "comprehensive", "research", "report"]):
                    intent = "deep_analysis"
                else:
                    intent = "simple_search"
        except Exception as e:
            intent = "simple_search"

        state["intent"] = intent
        state["messages"].append(f"Intent: {intent}")
        return state

    def extract_parameters(self, state: AgentState) -> AgentState:
        """Node 2: Use OPENAI to extract travel parameters"""
        system_prompt = """Extract travel search parameters from the user query.
        Return ONLY a valid JSON object with: origin, destination, date, passengers, check_in, check_out, guests.
        Default values: origin="Miami", destination="Madrid", date="2025-12-14", passengers=1,
                       check_in="2025-12-14", check_out="2025-12-17", guests=2

        Return ONLY the JSON, nothing else."""

        prompt = f"{system_prompt}\n\nUser query: {state['query']}\n\nJSON:"

        try:
            response = self.llm.invoke(prompt)
            content = response.content.strip()

            if "```json" in content:
                content = content.split("```json")[1].split("```")[0].strip()
            elif "```" in content:
                content = content.split("```")[1].split("```")[0].strip()

            params = json.loads(content)

            defaults = {
                "origin": "Miami",
                "destination": "Madrid",
                "date": "2025-12-14",
                "passengers": 1,
                "check_in": "2025-12-14",
                "check_out": "2025-12-17",
                "guests": 2
            }
            for key, default_val in defaults.items():
                if key not in params:
                    params[key] = default_val

        except:
            params = {
                "origin": "Miami",
                "destination": "Madrid",
                "date": "2025-12-14",
                "passengers": 1,
                "check_in": "2025-12-14",
                "check_out": "2025-12-17",
                "guests": 2
            }

        state["extracted_params"] = params
        state["messages"].append(f"Params extracted")
        return state

    def plan_tools(self, state: AgentState) -> AgentState:
        """Node 3: Decide which tools to use based on intent"""
        tool_plans = {
            "simple_search": ["search_flights"],
            "price_focused": ["search_flights", "analyze_prices"],
            "explore_alternatives": ["search_flights", "analyze_prices", "suggest_alternatives"],
            "deep_analysis": ["search_flights", "analyze_prices", "get_route_info", "suggest_alternatives"],
            "hotel_search": ["search_hotels", "analyze_hotel_prices"],
            "package_deal": ["search_flights", "search_hotels", "get_hotel_package"]
        }

        intent = state.get("intent", "simple_search")
        tools_to_use = tool_plans.get(intent, ["search_flights"])

        state["messages"].append(f"Planned: {len(tools_to_use)} tools")
        state["_tools_plan"] = tools_to_use
        return state

    def execute_tools(self, state: AgentState) -> AgentState:
        """Node 4: Execute ONLY the planned tools"""
        params = state["extracted_params"]
        tools_plan = state.get("_tools_plan", ["search_flights"])

        for tool_name in tools_plan:
            state["tools_used"].append(tool_name)

            if tool_name == "search_flights":
                state["flights"] = self.tools.search_flights(
                    params["origin"], params["destination"],
                    params["date"], params["passengers"]
                )
            elif tool_name == "analyze_prices":
                state["price_analysis"] = self.tools.analyze_prices(state["flights"])
            elif tool_name == "get_route_info":
                state["route_info"] = self.tools.get_route_info(
                    params["origin"], params["destination"]
                )
            elif tool_name == "suggest_alternatives":
                state["alternatives"] = self.tools.suggest_alternatives(
                    params["origin"], params["destination"]
                )
            elif tool_name == "search_hotels":
                state["hotels"] = self.tools.search_hotels(
                    params["destination"], params["check_in"],
                    params["check_out"], params["guests"]
                )
            elif tool_name == "analyze_hotel_prices":
                state["hotel_analysis"] = self.tools.analyze_hotel_prices(state["hotels"])
            elif tool_name == "get_hotel_package":
                state["package_info"] = self.tools.get_hotel_package(
                    params["origin"], params["destination"], params["date"],
                    params["check_in"], params["check_out"], params["passengers"]
                )

        state["messages"].append(f"Executed: {len(state['tools_used'])} tools")
        return state

    def consolidate_results(self, state: AgentState) -> AgentState:
        """Node 5: Use OPENAI to create final response"""
        context = f"Query: {state['query']}\nIntent: {state['intent']}\n"

        if state.get("flights"):
            context += f"Flights: {len(state['flights'])}"
        if state.get("hotels"):
            context += f"\nHotels: {len(state['hotels'])}"
        if state.get("package_info"):
            context += f"\nPackage: ${state['package_info'].get('total_package_price', 'N/A')}"

        try:
            response = self.llm.invoke(f"Summarize: {context}")
            state["final_response"] = response.content
        except:
            state["final_response"] = "Results found."

        return state

print(" Flight Agent with Hotel Support created!")
````

````python
from langgraph.graph import StateGraph, END

def create_flight_agent_graph(llm):
    """Create the LangGraph agent workflow"""
    agent = FlightAgent(llm)

    # Define the graph
    workflow = StateGraph(AgentState)

    # Add nodes (each step in the process)
    workflow.add_node("classify_intent", agent.classify_intent)
    workflow.add_node("extract_parameters", agent.extract_parameters)
    workflow.add_node("plan_tools", agent.plan_tools)
    workflow.add_node("execute_tools", agent.execute_tools)
    workflow.add_node("consolidate_results", agent.consolidate_results)

    # Define the flow (edges between nodes)
    workflow.set_entry_point("classify_intent")
    workflow.add_edge("classify_intent", "extract_parameters")
    workflow.add_edge("extract_parameters", "plan_tools")
    workflow.add_edge("plan_tools", "execute_tools")
    workflow.add_edge("execute_tools", "consolidate_results")
    workflow.add_edge("consolidate_results", END)

    # Compile the graph
    return workflow.compile()

# Create the agent
agent_graph = create_flight_agent_graph(llm)
print(" LangGraph agent compiled successfully!")
print(" Agent Flow: classify → extract → plan → execute → consolidate → END")
````

````python
from typing import TypedDict, List, Dict, Any
from langgraph.graph import StateGraph, END
import json

# Define a completely new state class
class TravelAgentState(TypedDict):
    """State for travel agent - fresh version"""
    query: str
    intent: str
    extracted_params: Dict[str, Any]
    tools_used: List[str]
    _tools_plan: List[str]
    flights: List[Dict[str, Any]]
    price_analysis: Dict[str, Any]
    route_info: Dict[str, Any]
    alternatives: Dict[str, Any]
    hotels: List[Dict[str, Any]]
    hotel_analysis: Dict[str, Any]
    package_info: Dict[str, Any]
    final_response: str
    messages: List[str]

print(" TravelAgentState defined!")


class TravelAgent:
    """Fixed agent without operator.add issues"""
    def __init__(self, llm):
        self.llm = llm
        self.tools = FlightTools()

    def classify_intent(self, state: TravelAgentState) -> TravelAgentState:
        """Node 1: Classify user intent"""
        system_prompt = """You are an intent classification expert for a travel booking system.
        Classify the user query into one of these intents:
        - simple_search: Basic flight search
        - price_focused: User wants price analysis and deals
        - explore_alternatives: User wants to see alternative options
        - deep_analysis: User wants comprehensive analysis with all information
        - hotel_search: User wants to search for hotels only
        - package_deal: User wants both flights and hotels together

        Respond with ONLY the intent name in lowercase, nothing else."""

        prompt = f"{system_prompt}\n\nUser query: {state['query']}\n\nIntent:"

        try:
            response = self.llm.invoke(prompt)
            intent = response.content.strip().lower().replace('.', '').replace(',', '')

            valid_intents = ["simple_search", "price_focused", "explore_alternatives",
                           "deep_analysis", "hotel_search", "package_deal"]

            if intent not in valid_intents:
                query_lower = state["query"].lower()
                if ("hotel" in query_lower or "accommodation" in query_lower) and "flight" not in query_lower:
                    intent = "hotel_search"
                elif "hotel" in query_lower and "flight" in query_lower:
                    intent = "package_deal"
                elif "cheap" in query_lower or "best price" in query_lower:
                    intent = "price_focused"
                elif "alternative" in query_lower or "nearby" in query_lower:
                    intent = "explore_alternatives"
                elif any(word in query_lower for word in ["analyze", "compare", "complete", "deep"]):
                    intent = "deep_analysis"
                else:
                    intent = "simple_search"
        except:
            intent = "simple_search"

        state["intent"] = intent
        state["messages"] = state.get("messages", []) + [f"Intent: {intent}"]
        return state

    def extract_parameters(self, state: TravelAgentState) -> TravelAgentState:
        """Node 2: Extract parameters"""
        system_prompt = """Extract travel search parameters from the user query.
        Return ONLY a valid JSON object with: origin, destination, date, passengers, check_in, check_out, guests.
        Default values: origin="Miami", destination="Madrid", date="2025-12-14", passengers=1,
                       check_in="2025-12-14", check_out="2025-12-17", guests=2

        Return ONLY the JSON, nothing else."""

        prompt = f"{system_prompt}\n\nUser query: {state['query']}\n\nJSON:"

        try:
            response = self.llm.invoke(prompt)
            content = response.content.strip()
            if "```json" in content:
                content = content.split("```json")[1].split("```")[0].strip()
            elif "```" in content:
                content = content.split("```")[1].split("```")[0].strip()
            params = json.loads(content)

            defaults = {
                "origin": "Miami", "destination": "Madrid", "date": "2025-12-14",
                "passengers": 1, "check_in": "2025-12-14", "check_out": "2025-12-17", "guests": 2
            }
            for key, val in defaults.items():
                if key not in params:
                    params[key] = val
        except:
            params = {
                "origin": "Miami", "destination": "Madrid", "date": "2025-12-14",
                "passengers": 1, "check_in": "2025-12-14", "check_out": "2025-12-17", "guests": 2
            }

        state["extracted_params"] = params
        state["messages"] = state.get("messages", []) + ["Params extracted"]
        return state

    def plan_tools(self, state: TravelAgentState) -> TravelAgentState:
        """Node 3: Plan which tools to use"""
        tool_plans = {
            "simple_search": ["search_flights"],
            "price_focused": ["search_flights", "analyze_prices"],
            "explore_alternatives": ["search_flights", "analyze_prices", "suggest_alternatives"],
            "deep_analysis": ["search_flights", "analyze_prices", "get_route_info", "suggest_alternatives"],
            "hotel_search": ["search_hotels", "analyze_hotel_prices"],
            "package_deal": ["search_flights", "search_hotels", "get_hotel_package"]
        }

        intent = state.get("intent", "simple_search")
        tools_to_use = tool_plans.get(intent, ["search_flights"])

        state["_tools_plan"] = tools_to_use
        state["messages"] = state.get("messages", []) + [f"Planned: {len(tools_to_use)} tools"]
        return state

    def execute_tools(self, state: TravelAgentState) -> TravelAgentState:
        """Node 4: Execute the planned tools"""
        params = state["extracted_params"]
        tools_plan = state.get("_tools_plan", ["search_flights"])

        # Initialize tools_used as empty list (not appending!)
        executed_tools = []

        for tool_name in tools_plan:
            executed_tools.append(tool_name)

            if tool_name == "search_flights":
                state["flights"] = self.tools.search_flights(
                    params["origin"], params["destination"], params["date"], params["passengers"]
                )
            elif tool_name == "analyze_prices":
                state["price_analysis"] = self.tools.analyze_prices(state["flights"])
            elif tool_name == "get_route_info":
                state["route_info"] = self.tools.get_route_info(params["origin"], params["destination"])
            elif tool_name == "suggest_alternatives":
                state["alternatives"] = self.tools.suggest_alternatives(params["origin"], params["destination"])
            elif tool_name == "search_hotels":
                state["hotels"] = self.tools.search_hotels(
                    params["destination"], params["check_in"], params["check_out"], params["guests"]
                )
            elif tool_name == "analyze_hotel_prices":
                state["hotel_analysis"] = self.tools.analyze_hotel_prices(state["hotels"])
            elif tool_name == "get_hotel_package":
                state["package_info"] = self.tools.get_hotel_package(
                    params["origin"], params["destination"], params["date"],
                    params["check_in"], params["check_out"], params["passengers"]
                )

        state["tools_used"] = executed_tools
        state["messages"] = state.get("messages", []) + [f"Executed: {len(executed_tools)} tools"]
        return state

    def consolidate_results(self, state: TravelAgentState) -> TravelAgentState:
        """Node 5: Create detailed final response"""
        return generate_enhanced_response(state)


def create_travel_agent_graph(llm):
    """Create the fixed workflow"""
    agent = TravelAgent(llm)
    workflow = StateGraph(TravelAgentState)

    workflow.add_node("classify_intent", agent.classify_intent)
    workflow.add_node("extract_parameters", agent.extract_parameters)
    workflow.add_node("plan_tools", agent.plan_tools)
    workflow.add_node("execute_tools", agent.execute_tools)
    workflow.add_node("consolidate_results", agent.consolidate_results)

    workflow.set_entry_point("classify_intent")
    workflow.add_edge("classify_intent", "extract_parameters")
    workflow.add_edge("extract_parameters", "plan_tools")
    workflow.add_edge("plan_tools", "execute_tools")
    workflow.add_edge("execute_tools", "consolidate_results")
    workflow.add_edge("consolidate_results", END)

    return workflow.compile()

# Create the NEW fixed agent
agent_graph = create_travel_agent_graph(llm)
print(" Fixed Travel Agent created!")
````

````python
def generate_enhanced_response(state: dict) -> dict:
    """
    Enhanced response generator that prominently displays hotels
    """
    response_parts = []

    query = state.get("query", "")
    intent = state.get("intent", "")
    params = state.get("extracted_params", {})

    # Greeting based on intent
    if intent == "hotel_search":
        response_parts.append(f"HOTEL SEARCH RESULTS FOR {params.get('destination', 'YOUR DESTINATION').upper()}\n\n")
    elif intent == "package_deal":
        response_parts.append(f"COMPLETE TRAVEL PACKAGE TO {params.get('destination', 'YOUR DESTINATION').upper()}\n\n")
    elif intent == "simple_search" or intent == "price_focused":
        response_parts.append(f"FLIGHT OPTIONS: {params.get('origin', '')} → {params.get('destination', '')}\n\n")
    else:
        response_parts.append(f"YOUR TRAVEL SEARCH RESULTS\n\n")

    # ============================================================
    # FLIGHTS SECTION (if available)
    # ============================================================
    if state.get("flights") and len(state["flights"]) > 0:
        response_parts.append("FLIGHT OPTIONS \n")
        response_parts.append("\n\n")

        # Show top 5 flights
        for i, flight in enumerate(state["flights"][:5], 1):
            departure = flight.get('departure_time', 'N/A')
            arrival = flight.get('arrival_time', 'N/A')
            duration = flight.get('duration', 'N/A')
            stops = flight.get('stops_text', flight.get('stops', 'N/A'))

            response_parts.append(
                f"{i}. {flight['airline']} - ${flight['price']}\n"
                f"   Departs: {departure} → Arrives: {arrival}\n"
                f"   Duration: {duration} | {stops}\n"
                f"   Book via: {flight.get('provider', 'N/A')}\n\n"
            )

        # Price analysis
        if state.get("price_analysis"):
            pa = state["price_analysis"]
            response_parts.append("Price Insights:\n")
            response_parts.append(f"   • Best Deal: ${pa.get('min', 'N/A')}\n")
            response_parts.append("\n\n")
            response_parts.append(f"   • Average: ${pa.get('average', 'N/A')}\n")
            response_parts.append("\n\n")
            response_parts.append(f"   • {pa.get('recommendation', '')}\n")
            if pa.get('best_time'):
                response_parts.append(f"   • {pa.get('best_time', '')}\n")
            response_parts.append("\n")

    # ============================================================
    # HOTELS SECTION (ENHANCED - PRIMARY FEATURE)
    # ============================================================
    if state.get("hotels") and len(state["hotels"]) > 0:

        response_parts.append("HOTEL RECOMMENDATIONS \n")


        # Show top 8 hotels with more details
        for i, hotel in enumerate(state["hotels"][:8], 1):
            amenities_list = hotel.get('amenities', [])
            amenities = ", ".join(amenities_list[:4])  # Show first 4 amenities
            if len(amenities_list) > 4:
                amenities += f", +{len(amenities_list) - 4} more"

            rating_stars = "⭐" * int(hotel.get('rating', 0))

            response_parts.append(
                f"{i}. {hotel['name']} {rating_stars} ({hotel.get('rating', 'N/A')})\n"
                f"   {hotel['price_per_night']}/night - {hotel.get('room_type', 'Standard Room')}\n"
                f"   {amenities}\n"
                f"   {hotel.get('distance_from_center', 'N/A')} from city center\n"
                f"   Book via: {hotel.get('provider', 'Direct Booking')}\n\n"
            )

        # Hotel analysis
        if state.get("hotel_analysis"):
            ha = state["hotel_analysis"]
            response_parts.append("Hotel Insights:\n")
            response_parts.append(f"   • Best Value: {ha.get('best_value', 'N/A')}\n")
            response_parts.append("\n\n")
            response_parts.append(f"   • Price Range: ${ha.get('min_price', 'N/A')} - ${ha.get('max_price', 'N/A')}\n")
            response_parts.append("\n\n")
            response_parts.append(f"   • Average Rating: {ha.get('avg_rating', 'N/A')}⭐\n")
            response_parts.append("\n\n")
            response_parts.append(f"   • {ha.get('recommendation', '')}\n\n")

    # ============================================================
    # PACKAGE DEAL (if both flights and hotels)
    # ============================================================
    if state.get("package_info") and state["package_info"].get("total_package_price"):
        pkg = state["package_info"]
        response_parts.append(" PACKAGE DEAL \n")
        response_parts.append("\n\n")
        response_parts.append(f"SPECIAL OFFER: Complete package for {pkg.get('total_package_price', 'N/A')}\n")
        response_parts.append(f"   • Flight: ${pkg.get('cheapest_flight_price', 'N/A')}\n")
        response_parts.append(f"   • Hotel: ${pkg.get('cheapest_hotel_per_night', 'N/A')}/night × {pkg.get('estimated_nights', 3)} nights\n")
        response_parts.append(f"   • {pkg.get('savings_tip', '')}\n\n")

    # ============================================================
    # ROUTE INFO
    # ============================================================
    if state.get("route_info"):
        ri = state["route_info"]
        response_parts.append("ℹTravel Information:\n")
        response_parts.append(f"   • Distance: {ri.get('distance', 'N/A')}\n")
        response_parts.append(f"   • Typical Flight Time: {ri.get('avg_duration', 'N/A')}\n")
        response_parts.append(f"   • Timezone: {ri.get('timezone', 'N/A')}\n")
        response_parts.append(f"   • Best Time to Visit: {ri.get('peak_season', 'N/A')}\n\n")

    # ============================================================
    # ALTERNATIVES
    # ============================================================
    if state.get("alternatives"):
        alt = state["alternatives"]
        response_parts.append("Money-Saving Tips:\n")
        if alt.get("nearby_destinations"):
            response_parts.append("   Consider nearby destinations:\n")
            for dest in alt["nearby_destinations"][:2]:
                response_parts.append(f"   • {dest}\n")
        if alt.get("alternate_dates"):
            response_parts.append("   Or try different dates:\n")
            for date_opt in alt["alternate_dates"][:2]:
                response_parts.append(f"   • {date_opt['date']}: Save {date_opt['savings']}\n")
        response_parts.append("\n")

    # ============================================================
    # CLOSING MESSAGE
    # ============================================================
    response_parts.append("\n  \n")
    response_parts.append("Need help with anything else? Just ask!\n")
    response_parts.append("   Try: \"Show me more hotels\" or \"Find package deals\"\n")

    final_response = "".join(response_parts)

    # Fallback if response is too short
    if len(final_response.strip()) < 100:
        final_response = (
            "I'm ready to help you with your travel plans! \n\n"
            "I can help you with:\n"
            "Flight searches\n"
            "Hotel recommendations\n"
            "What would you like to search for?"
        )

    state["final_response"] = final_response
    return state
````

````python
@app.route('/chat', methods=['POST'])
def chat():
    try:
        # Get the message
        data = request.get_json()
        user_message = data.get('message', '')

        print(f"💬 Received message: {user_message}")

        # Create state with query
        state = {
            "query": user_message,
            "intent": "",
            "extracted_params": {},
            "tools_used": [],
            "_tools_plan": [],
            "flights": [],
            "price_analysis": {},
            "route_info": {},
            "alternatives": {},
            "hotels": [],
            "hotel_analysis": {},
            "package_info": {},
            "final_response": "",
            "messages": []
        }

        print("🔄 Invoking agent...")

        # Invoke agent
        result = agent_graph.invoke(state)

        print("✅ Agent responded")

        # Get the final response (plain text with markdown)
        response_text = result.get('final_response', 'Sorry, I could not process that request.')

        return jsonify({
            'success': True,
            'response': response_text
        })

    except Exception as e:
        import traceback
        error_trace = traceback.format_exc()
        print(f"❌ ERROR:\n{error_trace}")

        return jsonify({
            'success': True,
            'response': f"Sorry, I encountered an error: {str(e)}\n\nPlease try the Quick Search form instead."
        })
````

````python
def generate_html_output(result):
    """Generate readable plain text output instead of HTML tags"""
    html = "<h2>Your Travel Results</h2>"

    # Display flights if available
    if result.get('flights'):
        html += "\n\nFLIGHT OPTIONS:\n"
        for flight in result['flights']:
            html += f"""
            Flight: {flight.get('airline', 'N/A')}
            Price: ${flight.get('price', 'N/A')}
            Duration: {flight.get('duration', 'N/A')}
            Departure: {flight.get('departure', 'N/A')}
            Arrival: {flight.get('arrival', 'N/A')}
            Stops: {flight.get('stops', 'N/A')}
            """

    # Display hotels if available
    if result.get('hotels'):
        html += "\n\nHOTEL RECOMMENDATIONS:\n"
        for hotel in result['hotels']:
            html += f"""
            Hotel: {hotel.get('name', 'N/A')} ({hotel.get('rating', 'N/A')}⭐)
            Price: ${hotel.get('price', 'N/A')} per night
            Amenities: {', '.join(hotel.get('amenities', []))}
            Distance: {hotel.get('distance', 'N/A')} km from city center
            """

    # Clean any leftover HTML tags
    clean_text_output = clean_html(html)
    return clean_text_output
````
